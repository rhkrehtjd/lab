- Ion Torrent sequencing Platforms
  - 반도체 칩을 활용
  - dna 합성 중에 발생하는 수소에 의한 PH 변화를 전기신호로 읽어서 sequencing 하는 방법
  - 반도체로 이루어진 칩 안의 여러개의 구멍에서 ..합성 발생
  - 원리
    - 분석 할 dna를 적당한 크기로 나누고, DNA의 양 말단에 증폭에 필요한 프라이머와 염기서열을 결합하여 라이브러리 제작
    - 라이브러리 분자들은 emulsion PCR 방법을 통해 DNA를 비드 표면에서 증폭하영 사용되며 한 개의 비드에는 한 가지 단일분자가 복제 증폭된 DNA가 결합
    - 
   - homopolymer error 발생가능
     - 한 염기가 연속적으로 분석되는 염기서열 부위
     

- application
  - 장비 안에 있는 반도체 칩의 버전에 따라 데이터 생산량이 달라짐
