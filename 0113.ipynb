> # 제 1강

- 생물정보학(bioinformatics)
  - 전산학, 통계학적 기술 $\to$ 생물학 문제 해결
  - stats + cs + biology
- 생물정보?
  - 그 정보의 양이 방대해짐
    - 이를테면, DNA를 이루는 염기 배열 순서나 단백질을 구성하는 아미노산 배열의 순서같은 것
    - 이러한 상당한 양의 생물정보를 처리 요구 $\uparrow$
    - 컴퓨터를 이용하여 체계적으로 분석 필요
- 생물정보학 활용
  - 게놈 프로젝트
  - 계통 조사 및 유사성 검색
    - 시퀀싱 $\to$ 염기서열 get 
    - 이 염기서열이 최후에 어떤 단백질로서 기능하는지 분석
    - 이 염기서열을 비교해서 어떤 특정 단백질의 기능이나 변이를 확인하는 일이 가능
    - 또, 특정 염기서열의 다른 종들과의 유사성에 대해 연구할 수 있음
  - 단백질 구조 규명
  - 유전자 상호작용 네트워크
    - 하나의 유전자가 하나의 기능만 하지 않음
      - 하나의 유전자가 여러 기능을 함, 다른 유전자들과도 긴밀한 관계를 맺음
      - 이런 네트워크들을 확인하여 이 유전자가 최종적으로 어떤 유전자들과 연관이 있고, 이 유전자들이 최종적으로 어떤 질병이나 기능들에 연관이 되어있는지 확인해야지 우리 몸에서 일어나는 시스템적인 부분들을 이해할 수 있음
      - 염기서열 하나를 읽어서 정보를 처리하지만 그 정보 자체는 여러 데이터베이스에서 쌓인 정보들을 통합적으로 분석해서 보는 것이기 때문에 통계나 수학, 컴퓨터 과학적 부분들이 동반되어야 함
   - 다중 오믹스 분석을 통한 바이오마커 발굴
     - 단계별로 나타날 수 있는 오믹스 데이터들을 가지고 통합분석을 하여 최종적으로 특정 질병의 발병이나 성질의 발현의 바이오 마커를 발굴
     - 이 데이터를 어떻게 연관하여 해석할지? $\rightarrow$ 생물정보학의 일종
   - 맞춤 의학 및 예방 의학
