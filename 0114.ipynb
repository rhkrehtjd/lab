- dna sequence analysis
  - sequencing으로 얻은 데이터는 굉장히 작은 조각임
  - 이렇게 작은 조각들의 상동성을 이용하여 연결 후 큰 서열을 만들어내는 게 목적
    - 이 과정은 assembly, 그 결과를 assembled sequence라고 함
  - 
  - assembly
    - De-novo
    - Reference (인간처럼 게놈과 같은 가이드가 있는 경우)

- basic steps of genome assembly
  - 

- 다중서열정렬(alignment)
  - 서열 간의 상관관계(상동성)를 알아보기 위해 DNA 또는 Protein 서열들을 정렬 알고리즘을 이용하여 배열하는 것
    - global alignment : 비교하는 서열 전체를 모두 비교해 유사도 측정
    - local alignment(활용도 높음) : 유사성 높은 각 서열의 부분만을 생각, 정확도는 좀 떨어질 수 있음
  - 관심대상인 하나의 서열과 상동성이 높은 서열들을 알아내어 서열의 기능 유추
  - 쌍서열정렬(2개의 서열비교), 다중서열정렬(3개 이상의 서열비교)
  


> alignment -> 유사성을 확인하는 것이기 때문에 계통에 대한 유연관계(생물들이 분류학적으로 얼마나 멀고 가까운지를 나타내는 관계) 분석 가능
> 실제로 유사성이 높으면 사실상 진화적인 부분에 있어 가깝게 있는 계통이라고 볼 수 있음

- 

- blast(Basic local alignment search tool)
  - 통계적으로 유효한 상동성을 찾을 목적으로 서열 데이터베이스를 검색하는 프로그램
  - alignment를 활용하여 만든 서열을 쌓여져 있는 데이터베이스에서 확인을 하여 유사 서열을 체크해볼 수 있음
