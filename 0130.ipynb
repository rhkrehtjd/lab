- sanger sequencing : 현재도 유전자 수준에서의 시퀀싱 방법으로서도 많이 활용되고 있음
  - dna 합성과정 이해 필요
  - 유튜브 이용해서 추가 공부 필요?..
  - dna 분자가 합성되는 과정에서 각 단계별로 사용되는 dntp종류를 선별할 수 있는 염기서열을 분석할 수 있게 됨
  - dna 분자 합성이 너무 빠르다보니 관찰할 방법 강구해야 함
  - ddntp 이용 - 하이드록시기가 없어서, 5'인산기가 반응하지 못하게 함, dna 중합효소가 ddntp를 만나게 되면 더 이상 합성이 불가능하게 됨, 연쇄반응이 정지됨
  - 따라서 네종류의 ddntp가 필요,,번거로움
  - 결론!
    - pcr과정에서 ddntp에 의해서 dna가 합성되지 않는 원리와 전기영동의 원리를 이용해서 염기서열을 확인하는 방법
    - 여전히 유전자 수준에서 긴 염기서열 분석을 위해서 사용되곤 함

- sanger 시퀀싱의 장단점
  - 장점
    - long reads
    - suitable for small projects
  - 단점
    - low throuput(적은 양의 서열 생산)
      - genome 하나를 sanger 시퀀싱으로 해결하려면 상당히 많은 비용과 시간이 듦
    - expensive

- sanger와 ngs의 비교
  - sanger $\to$ 정확도가 높지만, 특정 변이에 대한 후보들을 뽑아내기 위해선 sanger는 부적절
  - 
  - sanger는 한 샘플당 하나의 read, ngs는 샘플당 여러개의 read, 이렇게  throuput 같은 경우가 이렇게 차이남
  

- sanger vs ngs 상황별 선택
  - 제한된 수의 샘플이나, 특정 dna 작은 영역의 정보를 얻을 때는 sanger가 좋은 선택일 수 있음
  - 그 이상이지만, 전체 genome 수준이 아니라면 ngs에 있는 타겟 시퀀싱을 이용할 수도 있겠음
